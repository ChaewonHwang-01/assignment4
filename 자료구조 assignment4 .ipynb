#**1.LeetCode 225. Implement Stack using Queues**

```python
from collections import deque
class MyStack:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.__main_queue=deque()
        self.__sub_queue=deque()


    def push(self, x: int) -> None:
        """
        Push element x onto stack.
        """
        self.__main_queue.append(x) # Push to back .


    def pop(self) -> int:
        """
        Removes the element on top of the stack and returns that element.
        """
        ret=None
        while len(self.__main_queue)>1:
            temp=self.__main_queue.popleft() # Peek/pop from front .
            self.__sub_queue.append(temp) # Move main_queue to sub_queue

        if len(self.__main_queue) == 1:
            ret=self.__main_queue.popleft()
            self.__main_queue,self.__sub_queue=self.__sub_queue,self.__main_queue # Swap main_queue,sub_queue

        return ret



    def top(self) -> int:
        """
        Get the top element.
        """
        return self.__main_queue[-1]


    def empty(self) -> bool:
        """
        Returns whether the stack is empty.
        """
        return False if self.__main_queue else True



# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()
```

##풀이해설
이 문제는 큐(Queue)를 이용해서 스택(Stack)을 구현하는 문제다.  
스택은 후입선출(LIFO) 구조이고, 큐는 선입선출(FIFO) 구조이기 때문에, 큐의 동작을 조합해서 스택처럼 동작하도록 만들어야 한다.

#**2. LeetCode 232. Implement Queue using Stacks**

```python
class MyQueue {
public:
    stack<int> first, second;
    MyQueue() {

    }

    void push(int x) {
        first.push(x);
    }

    int pop() {
        while (first.size()) {
            second.push(first.top());
            first.pop();
        }
        int ret = second.top();
        second.pop();
        while (second.size()) {
            first.push(second.top());
            second.pop();
        }
        return ret;
    }

    int peek() {
        while (first.size()) {
            second.push(first.top());
            first.pop();
        }
        int ret = second.top();
        while (second.size()) {
            first.push(second.top());
            second.pop();
        }
        return ret;
    }

    bool empty() {
        if (first.size() == 0)
            return true;
        return false;
    }
};

/**
 * Your MyQueue object will be instantiated and called as such:
 * MyQueue* obj = new MyQueue();
 * obj->push(x);
 * int param_2 = obj->pop();
 * int param_3 = obj->peek();
 * bool param_4 = obj->empty();
 */
```

##풀이해설
이 문제는 스택(Stack)을 사용하여 큐(Queue)를 구현하는 문제다.  
큐는 선입선출(FIFO) 구조이고, 스택은 후입선출(LIFO) 구조이기 때문에,  
두 개의 스택을 이용해 큐처럼 동작하도록 만들어야 한다.

## 연습문제1번

```python
class ListQueue:
    def __init__(self):
        self.__queue = []

    def enqueue(self, x):
        self.__queue.insert(0, x)  # 맨 앞에 삽입 (큐의 tail)

    def dequeue(self):
        return self.__queue.pop()  # 맨 뒤에서 제거 (큐의 front)

    def front(self):
        return self.__queue[-1]  # 맨 뒤가 front

    def isEmpty(self) -> bool:
        return len(self.__queue) == 0

    def dequeueAll(self):
        self.__queue.clear()
```

##풀이해설
기존 구현에서는 리스트의 **맨 앞을 큐의 tail**, **맨 뒤를 큐의 front**로 간주했지만,  
이 문제에서는 **리스트의 맨 앞을 front**, **맨 뒤를 tail**로 바꾸어 구현해야 한다.

##연습문제2번

```python
from collections import deque

def is_in_the_language(s: str) -> bool:
    q1 = deque()
    q2 = deque()

    found_dollar = False
    for ch in s:
        if ch == '$':
            found_dollar = True
            continue
        if not found_dollar:
            q1.append(ch)  # w
        else:
            q2.appendleft(ch)  # w[::-1]

    # 두 큐가 동일한지 비교
    if len(q1) != len(q2):
        return False

    while q1:
        if q1.popleft() != q2.popleft():
            return False
    return True
```

##풀이해설
- `$` 이전의 문자들을 `q1`에 저장
- `$` 이후의 문자들을 `q2`에 **역순으로 저장** (`appendleft`)
- `q1`과 `q2`를 앞에서부터 하나씩 꺼내며 비교

##연습문제4번

```python
from collections import deque

class MyStack:
    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def push(self, x):
        self.q1.append(x)  # 그냥 큐에 추가 (O(1))

    def pop(self):
        # q1에서 마지막 요소를 남기고 모두 q2로 이동
        while len(self.q1) > 1:
            self.q2.append(self.q1.popleft())

        # 마지막 요소가 스택의 top → pop()으로 제거
        top = self.q1.popleft()

        # q1 <-> q2 swap
        self.q1, self.q2 = self.q2, self.q1

        return top
```

##풀이해설
- `push(x)`  
  → `q1` 큐에 그대로 추가 (O(1))

- `pop()`  
  → `q1`의 원소를 하나 남기고 모두 `q2`로 이동  
  → 마지막 원소(가장 나중에 추가된 것)가 스택의 top이므로 꺼냄  
  → `q1`과 `q2`를 스왑해서 다음 연산에 대비

##연습문제5번

```python
class MyQueue:
    def __init__(self):
        self.stack1 = []  # 입력용
        self.stack2 = []  # 출력용

    def enqueue(self, x):
        self.stack1.append(x)  # 스택1에 그대로 넣음

    def dequeue(self):
        while self.stack1:
            self.stack2.append(self.stack1.pop())  # 뒤집어서 스택2로
        ret = self.stack2.pop()  # 실제 큐의 front
        while self.stack2:
            self.stack1.append(self.stack2.pop())  # 다시 원상복구
        return ret
```

##풀이해설
- `enqueue(x)`  
  → 스택1에 그대로 `append(x)` (O(1))

- `dequeue()`  
  → 스택1의 모든 요소를 `pop()`하여 스택2에 `append()` (역순 생성)  
  → 스택2에서 `pop()`하여 큐의 front를 꺼냄  
  → 다시 스택2의 모든 요소를 스택1로 되돌림 (원상 복구)

##연습문제8번

```python
class Deque:
    def __init__(self):
        self.__queue = []

    def enqueueFront(self, x):
        self.__queue.insert(0, x)  # 맨 앞에 삽입

    def enqueueRear(self, x):
        self.__queue.append(x)  # 맨 뒤에 삽입

    def dequeueFront(self):
        if self.isEmpty():
            return None
        return self.__queue.pop(0)  # 맨 앞 제거

    def dequeueRear(self):
        if self.isEmpty():
            return None
        return self.__queue.pop()  # 맨 뒤 제거

    def front(self):
        if self.isEmpty():
            return None
        return self.__queue[0]

    def rear(self):
        if self.isEmpty():
            return None
        return self.__queue[-1]

    def isEmpty(self) -> bool:
        return len(self.__queue) == 0

    def dequeueAll(self):
        self.__queue.clear()

    def printQueue(self):
        print("Deque from front:", end=' ')
        for item in self.__queue:
            print(item, end=' ')
        print()
```
